# Write a python program to solve a constraint optimization  problem: Minimize/Maximize $f(x,y) = 81x^2 + y^2$ subject to $4x^2 + y^2  = 9$.

```python
import numpy as np
from scipy.optimize import minimize

# Define the objective function
def objective(x):
    x, y = x
    return 81 * x**2 + y**2

# Define the constraint function
def constraint(x):
    x, y = x
    return 4 * x**2 + y**2 - 9

# Initial guess for x and y
x0 = np.array([0.000, 0.01])

# Set up the optimization problem
constraints = {'type': 'eq', 'fun': constraint}
result = minimize(objective, x0, constraints=constraints)

# Print the results
print("Optimal values:")
print(f"x = {result.x[0]:.4f}")
print(f"y = {result.x[1]:.4f}")
print(f"Minimum value of f(x, y) = {result.fun:.4f}")
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the objective function f(x, y)
def f(x, y):
    return 81 * (x**2) + (y**2)

# Define the constraint function g(x, y)
def g(x, y):
    return 4 * (x**2) + (y**2)

# Create a meshgrid for x and y values
x_vals = np.linspace(-1.5, 1.5, 100)
y_vals = np.linspace(-3.5, 3.5, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z_f = f(X, Y)
Z_g = g(X, Y)

# Create the 3D surface plot
fig = go.Figure()

# Add the surface for f(x, y)
fig.add_trace(go.Surface(z=Z_f, x=X, y=Y, colorscale='Viridis', name='f(x, y)'))

# Add the constraint surface for g(x, y)
fig.add_trace(go.Surface(z=Z_g, x=X, y=Y, colorscale='Reds', opacity=0.5, showscale=False, name='4x^2 + y^2 = 9'))

# Set layout
fig.update_layout(
    title='3D Surface Plot',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='Value',
    ),
    margin=dict(l=0, r=0, b=0, t=40),
)

# Show the interactive plot
fig.show()
```

# Write a python program to solve a constraint optimization  problem: Minimize/Maximize $f(x,y) = 8x^2 - 2y$ subject to $x^2 + y^2  = 1$.
