# Define the Function to Optimize

$$
J(w_1) = (w_1 - 5)^2 + 10
$$

The objective is to find $w_1$ that minimizes $J(w_1)$.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Serif'
plt.rcParams['font.size'] = 15

# Define the cost function J(w_1)
def J(w_1):
  return (w_1 - 5)**2 + 10
```

# Initiailize $w_1$

```python
# Initialize w_1
w_1 = tf.Variable(0.0)
```

# Define Optimizer: we are going to using Stochastic Gradient Descent

```python
# Set up the optimizer
# We use the learning rate of 0.1
optimizer = tf.optimizers.SGD(learning_rate=0.1)
```

# Perform the Gradient Descent to get $w_1$ that Minimizes the Cost Function

```python
w_1_values = []
losses = []
# Perform gradient descent
for i in range(100):
  with tf.GradientTape() as tape:
    loss = J(w_1)
  gradients = tape.gradient(loss, [w_1])
  optimizer.apply_gradients(zip(gradients, [w_1]))
  w_1_values.append(w_1.numpy())
  losses.append(loss.numpy())
```

# Visualize the Loss as a function of iteration (also called epochs)


```python
# Plot the value of w_1 as a function of iterations
plt.plot(losses)
plt.xlabel('Iteration')
plt.ylabel('loss')
plt.title('Loss as a function of iterations')
```

# Visualize $w_1$ as a function of iteration

```python
# Plot the value of w_1 as a function of iterations
plt.plot(w_1_values)
plt.xlabel('Iteration')
plt.ylabel('$w_1$')
plt.title('Value of $w_1$ as a function of iterations')
```
